Before we start, just as a vaccination of sorts, the difference between numpy and lists

```python
# For math we'll be using numpy arrays

b = np.array([1,2,3])
print(b)

#Can we access an element of it?
print(b[1])   #Yes, it worked the same as a list

#Can we add a number to it?
print(b+10)

#Can you multiply by a number?
print(b*5)   #Yes ! And it actually multiplies every element !

#Can we add two arrays together?
print(b+b)   #Yes!!!! And even better, it adds element by element

#What if the vectors are of different size?
#print(b + np.array([1,2]))   #This would'nt work, because you cannot interpret it
```

# Birthday Paradox

* inside (), update your value for the total probability
* create np.array

```python
p = np.zeros(100)
p
```

```python
n = 365
p = np.zeros(n)  #probability of NOT having a matching birthday
#print(p)
p[0] = 1    # For 1 person, the prob of not having a matching a birthday is 1, haha, also p1 is one person
p[1] = 1 
#print(p[0])

#p[i] = 
for i in range(2,n):   # This loop will run from 2 to n-1 (so the highest number will be n-1)
        p[i]=(1-(n-1/365))*p[i-1]  #Calculate the probability that this new person doesn't have a matching birthday
        print(p[i])                   #Using this new value we just calculated, and a value p[i-1], calculate p[i]
            
#Plot p, which is a vector of numbers
```
